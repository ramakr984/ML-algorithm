```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
from math import sqrt

# Simulating AR(1) Process (for demonstration) with autocorrelation
np.random.seed(42)
n = 200  # length of time series
phi = 0.7  # AR(1) coefficient
epsilon = np.random.randn(n)  # Random noise (white noise)

# AR(1) process: Y_t = phi * Y_(t-1) + error_t
data = np.zeros(n)
for t in range(1, n):
    data[t] = phi * data[t-1] + epsilon[t]

# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data, label='Simulated AR(1) Data')
plt.title('Simulated AR(1) Process')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# ARIMA(p, d, q) model
# For demonstration, using ARIMA(1, 1, 1) (p=1, d=1, q=1)
# Note: d=1 for differencing to make series stationary

# Fit ARIMA model with order (p, d, q)
arima_model = sm.tsa.ARIMA(data, order=(1, 1, 1))  # ARIMA(1, 1, 1)
arima_model_fitted = arima_model.fit()

# Get the fitted values (predictions)
predictions = arima_model_fitted.predict(start=2, end=n-1)

# Plot the ACF and PACF plots to analyze autocorrelations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
plot_acf(data, lags=20, ax=ax1)
plot_pacf(data, lags=20, ax=ax2)
plt.suptitle('ACF and PACF Plots (Simulated AR(1) Process)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()

# Plot Original vs Predicted (from ARIMA model)
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data', color='blue')
plt.plot(range(2, n), predictions, label='Predicted Data', color='red', linestyle='--')
plt.title('Original vs Predicted Data (ARIMA(1,1,1))')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Compute RMSE (Root Mean Squared Error) between original and predicted data
# We compare the original values starting from time 2 since ARIMA(1,1,1) model needs differencing
rmse = sqrt(mean_squared_error(data[2:], predictions))
print(f'RMSE: {rmse}')
```
