```python


x=a[['id', 'age', 'height(cm)', 'weight(kg)', 'waist(cm)', 'eyesight(left)',
       'eyesight(right)', 'hearing(left)', 'hearing(right)', 'systolic',
       'relaxation', 'fasting blood sugar', 'Cholesterol', 'triglyceride',
       'HDL', 'LDL', 'hemoglobin', 'Urine protein', 'serum creatinine', 'AST',
       'ALT', 'Gtp', 'dental caries']]



y=a[['smoking']]



from sklearn.model_selection import train_test_split


X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.5,test_size=0.3)
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=100) 
```

```python
from sklearn.ensemble import RandomForestClassifier


rf=RandomForestClassifier(criterion='entropy',max_depth=12,max_features=1,max_leaf_nodes=13,random_state=10)


rf.fit(X_train,y_train)


```

```python
import pickle

with open('krishna.pickle','wb') as file:
    pickle.dump(rf,file)



```

```python

y_predict=rf.predict(X_train)

```

```python
l=rf.score(X_train,y_train)

print(l*100)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
# from sklearn.tree import DecisionTree
from sklearn.neighbors import KNeighborsClassifier


# we implemented the all the algorithm from here we used voting for the best fit
s=SVC(probability=True, kernel='linear', random_state=42 )
dt=DecisionTreeClassifier(random_state=24)
knn=KNeighborsClassifier(n_neighbors=5)


vc=VotingClassifier(estimators=[('s',s),('dt',dt),('knn',knn)],voting='soft',weights=[1,1,1])

vc.fit(X_train,y_train)
```

```python
with open('krishna.pickle','wb') as file:
    pickle.dump(vc,file)
```
