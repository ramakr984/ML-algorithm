```python
import pandas as pd


a=pd.read_csv(r"C:\Users\HP\Desktop\Loan.csv")
a
```

```python
a['ApplicationDate'] = pd.to_datetime(a['ApplicationDate'], format='mixed')

a['ApplicationDate'] = a['ApplicationDate'].apply(lambda x: x.toordinal())
```

```python
a['ApplicationDate'] = a['ApplicationDate'].astype(int)
```

```python
x = a[['ApplicationDate', 'Age', 'AnnualIncome', 'CreditScore',
       'Experience', 'LoanAmount',
       'LoanDuration', 'NumberOfDependents',
       'MonthlyDebtPayments',
       'CreditCardUtilizationRate', 'NumberOfOpenCreditLines',
       'NumberOfCreditInquiries', 'DebtToIncomeRatio', 'BankruptcyHistory',
       'PreviousLoanDefaults', 'PaymentHistory',
       'LengthOfCreditHistory', 'SavingsAccountBalance',
       'CheckingAccountBalance', 'TotalAssets', 'TotalLiabilities',
       'MonthlyIncome', 'UtilityBillsPaymentHistory', 'JobTenure', 'NetWorth',
       'BaseInterestRate', 'InterestRate', 'MonthlyLoanPayment',
       'TotalDebtToIncomeRatio', 'RiskScore']]
y = a[["LoanApproved"]]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
predict  = model.predict(x_test)

predict
```

```python
a = pd.DataFrame(predict,columns=["Loan"])
a
```

```python
from sklearn.tree import DecisionTreeRegressor


dr=DecisionTreeRegressor()


dr.fit(x_train,y_train)

hi3=dr.score(x_test,y_test)

DS3=dr.score(x_train,y_train)

print(DS3*100)
print(hi3*100)


dr.fit(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier


kc=KNeighborsClassifier()



kc.fit(x_train,y_train)

hi4=kc.score(x_test,y_test)

DS4=kc.score(x_train,y_train)

print(DS4*100)
print(hi4*100)


kc.fit(x_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB



nb=GaussianNB()


nb.fit(x_train,y_train)

hi5=nb.score(x_test,y_test)

DS5=nb.score(x_train,y_train)

print(DS5*100)
print(hi5*100)


nb.fit(x_train,y_train)

nb.fit(x_train,y_train)
```
