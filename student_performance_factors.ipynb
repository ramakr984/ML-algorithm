```python
# Separate categorical and numerical columns
categorical_columns = a.select_dtypes(include=['object']).columns
numerical_columns = a.select_dtypes(include=['number']).columns

categorical_data = a[categorical_columns]
numerical_data = a[numerical_columns]

print("Categorical Columns:\n", categorical_data)
print("\nNumerical Columns:\n", numerical_data)
```

```python
from sklearn.preprocessing import LabelEncoder
# Initialize the LabelEncoder
label_encoder = LabelEncoder()
# Apply LabelEncoder to each categorical column


# Apply LabelEncoder to each categorical column
for column in a.select_dtypes(include=['object']).columns:
    # Create a new LabelEncoder for each column
    label_encoder = LabelEncoder()
    a[column] = label_encoder.fit_transform(a[column])

print(a)

```
