```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string
import re
from nltk import ne_chunk, pos_tag
```

```python
# Input text
text = "The weather is sunny and pleasant. Natural Language Processing is fascinating!"

# Text Normalization: Lowercasing and removing punctuation
text_normalized = text.lower().translate(str.maketrans('', '', string.punctuation))

# Tokenization
tokens = word_tokenize(text_normalized)

stop_words = set(stopwords.words('english'))
tokens_without_stopwords = [word for word in tokens if word not in stop_words]

# Stemming
stemmer = PorterStemmer()
stems = [stemmer.stem(word) for word in tokens_without_stopwords]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmas = [lemmatizer.lemmatize(word) for word in tokens_without_stopwords]

# POS tagging
pos_tags = nltk.pos_tag(tokens)

# Output the POS tags
print(pos_tags)

print(tokens)
print(tokens_without_stopwords)
print(stems)
print(lemmas)
```

```python
import spacy

# Load spaCy's English model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "The weather is sunny and pleasant. Natural Language Processing is fascinating! India"

# Process text using spaCy
doc = nlp(text)

# Text Normalization: Lowercasing
text_normalized = text.lower()

# Tokenization
tokens = [token.text for token in doc]

# Removing stopwords
tokens_without_stopwords = [token.text for token in doc if not token.is_stop]

# Lemmatization
lemmas = [token.lemma_ for token in doc]

# POS Tagging
pos_tags = [(token.text, token.pos_) for token in doc]

# Named Entity Recognition (NER)
entities = [(ent.text, ent.label_) for ent in doc.ents]

# Dependency Parsing
dependencies = [(token.text, token.dep_, token.head.text) for token in doc]

# Display results
print("Original Text:", text)
print("Normalized Text:", text_normalized)
print("Tokens:", tokens)
print("Tokens without Stopwords:", tokens_without_stopwords)
print("Lemmas:", lemmas)
print("POS Tags:", pos_tags)
print("Named Entities:", entities)
print("dependencies:", dependencies)
```
