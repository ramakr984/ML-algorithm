```python
import seaborn as sns
#internal quartile range to remove the outer layers in the box plot
q1=a['Newspaper'].quantile(0.25)
q3=a['Newspaper'].quantile(0.75)

IQR=q3-q1
lower_bound=q1- (IQR*1.5)
upper_bound=q3- (IQR*1.5)

for Newspaper in a['Newspaper']:
    if(Newspaper<lower_bound):
       print("lower bound")
    elif(Newspaper>upper_bound):
       print("upper bound")
```
